```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# =========================
# 2) Convertir les colonnes dates
# =========================
date_cols = [
    "order_purchase_timestamp",
    "order_delivered_customer_date",
    "order_estimated_delivery_date",
]
for c in date_cols:
    orders[c] = pd.to_datetime(orders[c], errors="coerce")

# =========================
# 3) Garder uniquement les commandes livrées + dates valides
# =========================
orders_del = orders[orders["order_status"] == "delivered"].copy()
orders_del = orders_del.dropna(subset=["order_purchase_timestamp", "order_delivered_customer_date"])

# =========================
# 4) Calculer les métriques délai
# =========================
# Délai réel de livraison (en jours)
orders_del["delivery_time_days"] = (
    orders_del["order_delivered_customer_date"] - orders_del["order_purchase_timestamp"]
).dt.total_seconds() / 86400

# Retard par rapport à la date estimée (en jours) -> positif = retard
orders_del["delay_days"] = (
    orders_del["order_delivered_customer_date"] - orders_del["order_estimated_delivery_date"]
).dt.total_seconds() / 86400

orders_del["is_late"] = orders_del["delay_days"] > 0

# (option) enlever les valeurs extrêmes pour mieux voir le graphe
orders_del = orders_del[(orders_del["delivery_time_days"] >= 0) & (orders_del["delivery_time_days"] <= 200)]

# =========================
# 5) KPIs rapides
# =========================
avg_delivery = orders_del["delivery_time_days"].mean()
median_delivery = orders_del["delivery_time_days"].median()
pct_late = orders_del["is_late"].mean() * 100
avg_delay = orders_del["delay_days"].mean()

print("=== KPIs Livraison ===")
print(f"Délai moyen (jours)        : {avg_delivery:.2f}")
print(f"Délai médian (jours)       : {median_delivery:.2f}")
print(f"% livraisons en retard     : {pct_late:.2f}%")
print(f"Retard moyen vs estimation : {avg_delay:.2f} jours")

# =========================
# 6) Graph 1 — Distribution du délai de livraison
# =========================
plt.figure(figsize=(12,5))
sns.histplot(orders_del["delivery_time_days"], bins=40, kde=True)
plt.title("Distribution du délai de livraison (jours)")
plt.xlabel("Délai de livraison (jours)")
plt.ylabel("Nombre de commandes")
plt.tight_layout()
plt.show()

# =========================
# 7) Graph 2 — % retards par mois (saisonnalité des retards)
# =========================
orders_del["order_month"] = orders_del["order_purchase_timestamp"].dt.to_period("M").dt.to_timestamp()

late_by_month = (
    orders_del.groupby("order_month")["is_late"]
    .mean()
    .reset_index(name="pct_late")
)
late_by_month["pct_late"] *= 100

plt.figure(figsize=(12,5))
plt.bar(late_by_month["order_month"], late_by_month["pct_late"])
plt.title("% de livraisons en retard par mois")
plt.xlabel("Mois")
plt.ylabel("% en retard")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# =========================
# 8) Graph 3 — Délai moyen vs estimation (delay_days)
# =========================
plt.figure(figsize=(12,5))
sns.histplot(orders_del["delay_days"].dropna(), bins=40, kde=True)
plt.title("Distribution du retard vs date estimée (jours)")
plt.xlabel("Delay (jours) : positif = retard, négatif = en avance")
plt.ylabel("Nombre de commandes")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


orders["order_purchase_timestamp"] = pd.to_datetime(
    orders["order_purchase_timestamp"]
)

# =========================
# 3) Créer la colonne mois
# =========================
orders["order_month"] = orders["order_purchase_timestamp"].dt.to_period("M").dt.to_timestamp()

# =========================
# 4) Nombre de commandes par mois
# =========================
monthly_orders = (
    orders.groupby("order_month")["order_id"]
    .nunique()
    .reset_index(name="nb_orders")
    .sort_values("order_month")
)

# =========================
# 5) Visualisation
# =========================
plt.figure(figsize=(12,5))
sns.lineplot(
    x=monthly_orders["order_month"].astype(str),
    y=monthly_orders["nb_orders"],
    marker="o"
)

plt.title("Nombre de commandes par mois")
plt.xlabel("Mois")
plt.ylabel("Nombre de commandes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
